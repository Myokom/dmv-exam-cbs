```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Calculate the duration of each event in days
events_df['Duration'] = (events_df['End Date'] - events_df['Start Date']).dt.days + 1

# Set up the figure
plt.figure(figsize=(12, 8))

# Plot each event as a horizontal line or point
for i, row in events_df.iterrows():
    start_date = row['Start Date']
    end_date = row['End Date']
    y_position = i  # Use the row index for the y-axis
    plt.plot([start_date, end_date], [y_position, y_position], 'b-', linewidth=2, label=None if i else 'Event Duration')
    plt.scatter(start_date, y_position, color='green', label=None if i else 'Start Date')
    plt.scatter(end_date, y_position, color='red', label=None if i else 'End Date')

# Customize the plot
plt.yticks(range(len(events_df)), events_df['Event Name'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
plt.gcf().autofmt_xdate()
plt.title('Event Distribution Over Time')
plt.xlabel('Date')
plt.ylabel('Events')
plt.grid(axis='x')
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
