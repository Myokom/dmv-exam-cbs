# Scatterplot (and how to avoid overplotting)

## Transparency

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.randn(500)
y = np.random.randn(500)

# Scatter plot with transparency (alpha)
plt.scatter(x, y, alpha=0.3)
plt.title("Scatter Plot with Transparency")
plt.show()
```

## Jittering

```python
import seaborn as sns

# Sample data
data = {'Category': ['A']*50 + ['B']*50, 'Value': np.random.randint(1, 10, 100)}

# Create a DataFrame
import pandas as pd
df = pd.DataFrame(data)

# Plot with jitter
sns.stripplot(x="Category", y="Value", data=df, jitter=True)
plt.title("Jittering to Reduce Overplotting")
plt.show()
```

## Aggregation

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.colorbar(label='Count')
plt.title("Hexbin Plot (Aggregation)")
plt.show()
```

# Heatmaps

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate sample data
np.random.seed(42)
data = np.random.rand(10, 10)  # 10x10 matrix of random values

# Create a DataFrame for better visualization
df = pd.DataFrame(data, columns=[f"Col_{i}" for i in range(1, 11)])

# Create heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Heatmap Example")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset with missing values
df = pd.DataFrame({
    'A': [1, np.nan, 3, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, 2, np.nan, 4, np.nan]
})

# Heatmap for missing data
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.show()
```

```python
import seaborn as sns

# Load sample dataset
df = sns.load_dataset("penguins")

# Compute the correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
# Convert the price column to numeric, removing any non-numeric characters
listings['price'] = listings['price'].replace('[\$,]', '', regex=True).astype(float)

# Create a boxplot for the price column with enhanced aesthetics and transparent background
plt.figure(figsize=(12, 8))                             # Set the figure size
sns.boxplot(x=listings['price'], color='skyblue')       # Create a boxplot with skyblue color
plt.title('Boxplot of Listing Prices', fontsize=16)     # Set the title with font size 16
plt.xlabel('Price', fontsize=14)                        # Set the x-axis label with font size 14
plt.xticks(fontsize=12)                                 # Set the x-axis tick labels font size to 12
plt.yticks(fontsize=12)                                 # Set the y-axis tick labels font size to 12
plt.grid(True, linestyle='--', alpha=0.7)               # Add a grid with dashed lines and 70% transparency
plt.gca().patch.set_alpha(0)                            # Set the background to be transparent
plt.show()                                              # Display the plot
```

## IQR

```python

# Assume df is your dataset and 'price' is the column with listing prices
Q1 =listings['price'].quantile(0.25)
Q3 =listings['price'].quantile(0.75)
IQR = Q3 - Q1

# Calculate upper bound
upper_bound = Q3 + 10 * IQR

# Remove only upper outliers
listings_cleaned = listings[listings['price'] <= upper_bound]

# Display percentage of data removed
removed_percentage = (1 - len(listings_cleaned) / len(listings)) * 100
print(f"Percentage of data removed: {removed_percentage:.2f}%")
```

```python

# Create a boxplot for the price column
plt.figure(figsize=(10, 6))
sns.boxplot(x=listings_cleaned['price'])
plt.title('Boxplot of Listing Prices')
plt.xlabel('Price')
plt.show()
```

# Faceted

## Example 1: Faceted Scatter Plot


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset("penguins")

# Facet Scatter Plot (Trellis)
g = sns.FacetGrid(df, col="species")  # Faceting by species
g.map(sns.scatterplot, "bill_length_mm", "bill_depth_mm")  # Scatter plot

plt.show()
```

## Example 2: Faceted Histogram

```python
g = sns.FacetGrid(df, col="species", sharex=True, sharey=False)
g.map(sns.histplot, "flipper_length_mm")

plt.show()
```

## Example 3: Faceted Bar Chart

```python
g = sns.FacetGrid(df, col="species", sharex=True, sharey=True)
g.map(sns.barplot, "island", "body_mass_g")

plt.show()
```

# Sankey

```python
import plotly.graph_objects as go

# Define the Sankey Diagram data
fig = go.Figure(go.Sankey(
    node=dict(label=["Website Visit", "Sign-Up", "Purchase", "Drop-Off"]),
    link=dict(
        source=[0, 1, 1],  # From Website Visit, Sign-Up
        target=[1, 2, 3],  # To Sign-Up, Purchase, Drop-Off
        value=[10000, 5000, 2000]  # Number of users at each step
    )
))

fig.show()
```
